```python
import pandas as pd
import torch
import torch.nn as nn
import numpy as np
```

```python
# 1. Define FeedForward model (must MATCH training code)
# -----------------------------
class FeedForward(nn.Module):
    def __init__(self, d_model, dropout=0.5):
        super().__init__()
        self.linear_1 = nn.Linear(d_model, d_model)
        self.dropout_1 = nn.Dropout(dropout)
        self.linear_2 = nn.Linear(d_model, 1)

    def forward(self, x):
        x = self.linear_1(x)
        x = torch.tanh(x)
        x = self.dropout_1(x)
        x = self.linear_2(x)
        return x
```

```python
# 2. Load NN_Input_Data.csv
# -----------------------------
input_path = "NN_Input_Data.csv"
df = pd.read_csv(input_path)
# Remove the first index-like column (Unnamed: 0 or similar)
X = df.drop(columns=[df.columns[0]]).values.astype("float32")
```

```python
# 3. Load trained model1.pt
# -----------------------------
model_path = "model1.pt"
model = FeedForward(d_model=12, dropout=0.5)

# Load model weights
state_dict = torch.load(model_path, map_location="cpu")
model.load_state_dict(state_dict)
model.eval()
```

```python

# -----------------------------
# 4. Predict Y
# -----------------------------
X_tensor = torch.tensor(X, dtype=torch.float32)

with torch.no_grad():
    preds = model(X_tensor).numpy().flatten()

# Y cannot be negative → clip
preds = preds.clip(min=0)
```

```python

# 5. Replace first column with Y_pred
# -----------------------------
df_out = df.copy()
df_out[df.columns[0]] = preds
df_out.rename(columns={df.columns[0]: "Y_predicted"}, inplace=True)
```
