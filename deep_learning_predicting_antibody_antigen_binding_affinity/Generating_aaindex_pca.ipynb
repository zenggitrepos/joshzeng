```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
 # Amino acids in standard order
AA = list("ARNDCQEGHILKMFPSTWYV")
```

```python
aaindex_df = pd.read_csv('datasets/aaindex1.csv', index_col=  0)
```

```python
aaindex_df.shape
```

```python
aaindex_df.iloc[0:2, 0:9]
```

```python
# Normalize each column
scaler = StandardScaler()
scaled = scaler.fit_transform(aaindex_df.T).T  # shape (20, n_features)
scaled.shape
```

```python
# Perform PCA (reduce to 20 components, adjustable)
pca = PCA(n_components=20)
pca_features = pca.fit_transform(scaled)

# Save to CSV
pca_df = pd.DataFrame(pca_features, index=AA,
                      columns=[f"PC{i+1}" for i in range(pca_features.shape[1])])
pca_df.reset_index(inplace=True)
pca_df.rename(columns={'index': 'AA'}, inplace=True)
```

```python

pca_df.to_csv("datasets/aaindex_pca.csv", index=False)
```
