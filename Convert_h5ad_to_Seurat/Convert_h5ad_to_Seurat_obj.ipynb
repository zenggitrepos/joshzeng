Convert 5had to Seurat


```python
import os
import scanpy as sc
from scipy import io
```

```python
file_name = 'TS_Skin.h5ad'
data_dir = '/media/josh/seagate/data/perm/scRNA_seq/human_refs'
sub_dir = file_name.replace('.h5ad', '')
dest_dir = f'{data_dir}/{sub_dir}'
if not os.path.isdir(dest_dir):
    os.makedirs(dest_dir)
```

```python
adata = sc.read_h5ad(f'{data_dir}/{file_name}')
adata = adata.raw.to_adata() 
if not os.path.isdir(f'{dest_dir}/matrix_files'):
    os.makedirs(f'{dest_dir}/matrix_files')
    
with open(f'{dest_dir}/matrix_files/barcodes.tsv', 'w') as f:
    for item in adata.obs_names:
        f.write(item + '\n')
        
with open(f'{dest_dir}/matrix_files/features.tsv', 'w') as f:
    for item in ['\t'.join([x,x,'Gene Expression']) for x in adata.var_names]:
        f.write(item + '\n')

io.mmwrite(f'{dest_dir}/matrix_files/matrix', adata.X.T)

!gzip /media/joshu/seagate/data/perm/scRNA_seq/human_refs/TS_Skin/matrix_files/*

adata.obs.to_csv(f'{dest_dir}/matrix_files/metadata.csv')
```
